```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
parkinsons_data = pd.read_csv('parkinsons.csv')
parkinsons_data.head()
```

```python
parkinsons_data.groupby('status').mean(numeric_only=True)
```

```python
#Data Preprocessing
X = parkinsons_data.drop(columns=['name', 'MDVP:Fo(Hz)' , 'MDVP:Jitter(Abs)', 'MDVP:RAP', 'MDVP:PPQ', 'Jitter:DDP', 'MDVP:Shimmer', 'Shimmer:DDA', 'NHR', 'HNR', 'status',	'RPDE',	'DFA', 'spread1', 'spread2', 'D2','PPE', 'MDVP:Shimmer(dB)', 'Shimmer:APQ3', 'Shimmer:APQ5', 'MDVP:APQ'
], axis=1)
Y = parkinsons_data['status']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_test.shape, X_train.shape)
```

```python
# Data Standarlisation
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)

X_test = scaler.transform(X_test)
```

```python
print(X_train)
```

```python
#Model Training
model = svm.SVC(kernel='linear')
```

```python
model.fit(X_train, y_train)
```

```python
# Accuracy score on training data

X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(y_train, X_train_prediction)
```

```python
print("Accuracy score of training data: ", training_data_accuracy)
```

```python
# Accuracy score on testing data

X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(y_test, X_test_prediction)
```

```python
print("Accuracy score of testing data: ", testing_data_accuracy)
```

Building a predictive system

```python
import numpy as np

input_data = (157.302, 74.997, 0.00784)

# changing input data into numpy array 
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

# standardize the data
std_data = scaler.transform(input_data_reshaped)

# make a prediction
prediction = model.predict(std_data)
print(prediction)

if(prediction[0] == 0):
    print("Person is not having Parkinsons Disease")
    
else:
    print("Person is having Parkinsons Disease")
```
